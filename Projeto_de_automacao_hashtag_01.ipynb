```python
import pyautogui
import pyperclip
import time

pyautogui.PAUSE = 1

# pyautogui.click > clicar
# pyautogui.press > aperta 1 tecla
# pyautoguihotkey > aperta um conjunto de teclas
# pyautogui.write > escreve um texto


# Passo 1: Entrar no sistema da empresa (Nesse caso, é o link do drive)
pyautogui.hotkey('ctrl', 't')
pyperclip.copy('https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga')
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')
time.sleep(3)

# Passo 2: Navegar no sistema e encontrar a base de dados (Entrar na pasta exportar)
pyautogui.click(x=346, y=311, clicks=2)
time.sleep(2) # opcional


# Passo 3: Fazer o download da base de vendas
pyautogui.click(x=357, y=397)
pyautogui.click(x=1160, y=203)
pyautogui.click(x=934, y=603)
time.sleep(3)

```

```python
# Passo 4: Importar a base de vendas para o python
import pandas as pd

tabela = pd.read_excel(r'C:\Users\Fabio\Downloads\Vendas - Dez.xlsx')
display(tabela)
```

```python
# Passo 5: Calcular os indicadores da impresa
faturamento = tabela['Valor Final'].sum()
print(faturamento)
quantidade = tabela['Quantidade'].sum()
print(quantidade)
```

```python
# Passo 6: Enviar um e-mail para a diretória com os indicadores de vendas

# Abre aba
pyautogui.hotkey('ctrl', 't')
# Entrar no link do e-mail (https://mail.google.com/mail/u/1/#inbox)
pyperclip.copy('https://mail.google.com/mail/u/1/#inbox')
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')
time.sleep(5)

# Clicar no botão escrever
pyautogui.click(x=81, y=212)
time.sleep(3)

# Preencher as informações do e-mail
pyautogui.write('fabiooliveiraideia2@gmail.com')
pyautogui.press('tab') # Selecionar o e-mail
pyautogui.press('tab') # Pular para o campo de assunto

pyperclip.copy('Relatório de vendas do dia anterior')
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('tab') # Pular para o campo de corpo do e-mail

texto = f'''
prezados,

Segue aqui o relatório de vendas do dia anterior.
Faturamento: R$ {faturamento:,.2f}
Quantidade de produtos vendidos: {quantidade:,.2f} Unidades

Qualquer dúvida fico à disposição. Abraço.
Att. Fabio Oliveira, engenheiro de software.
'''

pyperclip.copy(texto)
pyautogui.hotkey('ctrl', 'v')
time.sleep(2)

# Anexar um arquivo
arquivo = r'C:\Users\Fabio\Downloads\Vendas - Dez.xlsx'
pyautogui.click(x=959, y=695)
time.sleep(2)
pyperclip.copy(arquivo)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')

time.sleep(5)
# Clicar no botão enviar

# Apertar ctrl enter - Enviar o e-mail
pyautogui.hotkey('ctrl', 'enter')
```
